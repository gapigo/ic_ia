```python
import pandas as pd
import re
from copy import deepcopy
len(str_pandemias.split('\n\n'))
len(str_endemias.split('\n\n'))

def exclude_links(_str: str):
    pattern=r'(http|ftp|https):\/\/([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:\/~+#-]*[\w@?^=%&\/~+#-])'
    _res = deepcopy(_str)
    for found in re.finditer(pattern, _str):
        _res = _res.replace(str(found.group()), '')
    return _res

def get_info(doenca: str, title=False):
    rows = re.split(r'\n-', doenca)
    nome_doenca = rows[0]
    for _ in range(5):
        nome_doenca = nome_doenca.strip('\n').strip('.').strip()
    if title:
        return ('nome_doenca', *[info.split(': ')[0].strip('\n') for i, info in enumerate(rows) if i not in [0]])
    return (nome_doenca, *[info.split(': ')[1].strip('\n') for i, info in enumerate(rows) if i not in [0]])
pandemias = str_pandemias.split('\n\n')
pandemias = [exclude_links(doenca) for doenca in pandemias]
endemias = str_endemias.split('\n\n')


# pandemias[0]
# exclude_links(pandemias[9])
re.findall(r'^- .*$', pandemias[8])
print(pandemias[8].split('- ')[0].strip('\n').strip('.').strip())
# list_data = [get_info(doenca) for doenca in pandemias]
list_data = [get_info(doenca) for doenca in pandemias] + [get_info(doenca) for doenca in endemias]
pd.DataFrame(list_data, columns=get_info(pandemias[0], title=True))
```
